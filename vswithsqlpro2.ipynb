```python
import pandas as pd
import psycopg2

# Database connection parameters
DB_NAME = "project2db"
DB_USER = "postgres"
DB_PASSWORD = "V@risu18"
DB_HOST = "localhost"
DB_PORT = "5432"

# Load the CSV file
file_path = "C:\PARTHI\PROJECT - 2\Customers_Modified.csv"
df = pd.read_csv(file_path)

# Convert Birthday column to proper date format
df["Birthday"] = pd.to_datetime(df["Birthday"], errors="coerce")

# Establish connection to PostgreSQL
try:
    conn = psycopg2.connect(
        dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT
    )
    cursor = conn.cursor()
    print("Database connection successful!")

    # Create table with the appropriate schema
    create_table_query = """
    CREATE TABLE IF NOT EXISTS customers (
        CustomerKey INT PRIMARY KEY,
        Gender VARCHAR(10),
        Name VARCHAR(100),
        City VARCHAR(100),
        State VARCHAR(100),
        Zip_Code VARCHAR(20),
        Country VARCHAR(100),
        Continent VARCHAR(50),
        Birthday DATE
    );
    """
    cursor.execute(create_table_query)
    conn.commit()
    print("Table created successfully!")

    # Insert data into the table
    for _, row in df.iterrows():
        insert_query = """
        INSERT INTO customers (CustomerKey, Gender, Name, City, State, Zip_Code, Country, Continent, Birthday)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (CustomerKey) DO NOTHING;
        """
        cursor.execute(
            insert_query,
            (
                row["CustomerKey"],
                row["Gender"],
                row["Name"],
                row["City"],
                row["State"],
                row["Zip Code"],
                row["Country"],
                row["Continent"],
                row["Birthday"],
            ),
        )

    conn.commit()
    print("Data inserted successfully!")

except Exception as e:
    print("Error:", e)

finally:
    if cursor:
        cursor.close()
    if conn:
        conn.close()
    print("Database connection closed.")
```

```python
import psycopg2
import pandas as pd

# Database connection details
DB_NAME = "project2db"
DB_USER = "postgres"
DB_PASSWORD = "V@risu18"
DB_HOST = "localhost"  # or your database server IP
DB_PORT = "5432"  # Default PostgreSQL port

# CSV file path
csv_file_pathc = "C:\PARTHI\PROJECT - 2\Exchange_Rates_Cleaned.csv"

# Load CSV data
df = pd.read_csv(csv_file_pathc)

# Convert Date column to proper format
df["Date"] = pd.to_datetime(df["Date"])

# Establish connection to PostgreSQL
try:
    conn = psycopg2.connect(
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD,
        host=DB_HOST,
        port=DB_PORT
    )
    cursor = conn.cursor()
    print("Database connection successful!")
except Exception as e:
    print(f"Error connecting to database: {e}")
    exit()

# Create Exchange Rates table
create_table_query = """
CREATE TABLE IF NOT EXISTS exchange_rates (
    id SERIAL PRIMARY KEY,
    date TIMESTAMP,
    currency VARCHAR(10),
    exchange DECIMAL(10,4)
);
"""

try:
    cursor.execute(create_table_query)
    conn.commit()
    print("Table 'exchange_rates' created successfully!")
except Exception as e:
    print(f"Error creating table: {e}")
    conn.rollback()

# Insert data into exchange_rates table
insert_query = """
INSERT INTO exchange_rates (date, currency, exchange)
VALUES (%s, %s, %s);
"""

data = df.values.tolist()

try:
    cursor.executemany(insert_query, data)
    conn.commit()
    print("Data inserted successfully into 'exchange_rates'!")
except Exception as e:
    print(f"Error inserting data: {e}")
    conn.rollback()

# Closing the connection
cursor.close()
conn.close()
print("Database connection closed.")
```

```python
import psycopg2
import pandas as pd

# Database connection details
DB_NAME = "project2db"
DB_USER = "postgres"
DB_PASSWORD = "V@risu18"
DB_HOST = "localhost"  # or your database server IP
DB_PORT = "5432"  # Default PostgreSQL port

# CSV file path
csv_file_pathf = "C:\PARTHI\PROJECT - 2\Stores_Cleaned.csv"

# Load CSV data
df = pd.read_csv(csv_file_pathf, encoding="ISO-8859-1")

# Convert date column to proper format
df["Open Date"] = pd.to_datetime(df["Open Date"])

# Establish connection to PostgreSQL
try:
    conn = psycopg2.connect(
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD,
        host=DB_HOST,
        port=DB_PORT
    )
    cursor = conn.cursor()
    print("Database connection successful!")
except Exception as e:
    print(f"Error connecting to database: {e}")
    exit()

# Create Stores table
create_table_query = """
CREATE TABLE IF NOT EXISTS stores (
    store_key INTEGER PRIMARY KEY,
    country VARCHAR(255),
    state VARCHAR(255),
    square_meters FLOAT,
    open_date TIMESTAMP
);
"""

try:
    cursor.execute(create_table_query)
    conn.commit()
    print("Table 'stores' created successfully!")
except Exception as e:
    print(f"Error creating table: {e}")
    conn.rollback()

# Insert data into stores table
insert_query = """
INSERT INTO stores (store_key, country, state, square_meters, open_date)
VALUES (%s, %s, %s, %s, %s)
ON CONFLICT (store_key) DO NOTHING;
"""

# Convert DataFrame to list of tuples for insertion
data = df.values.tolist()

try:
    cursor.executemany(insert_query, data)
    conn.commit()
    print("Data inserted successfully into 'stores'!")
except Exception as e:
    print(f"Error inserting data: {e}")
    conn.rollback()

# Closing the connection
cursor.close()
conn.close()
print("Database connection closed.")
```
